# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## School Summary

```python
# Use the code provided to select the school type
school_types = school_data.set_index(["school_name"])["type"].reset_index(name = 'school_type')
school_types
```

```python
# Calculate the total student count
per_school_counts = school_data_complete.groupby(["school_name"]).aggregate({'school_name':'count'})
per_school_counts.columns = ['total_students']
per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_student  = (student_data.groupby('school_name')['Student ID'].count().reset_index())

per_school_budget = (school_data.groupby('school_name')['budget'].sum().reset_index())

per_school_capita = per_school_budget[['school_name','budget']]

per_school_capita['total_students']=per_school_student['Student ID']
per_school_capita['per_capita'] = per_school_capita['budget']/per_school_capita['total_students']
per_school_capita

```

```python
# Calculate the average test scores
avg_test_scores_df = school_data_complete.groupby("school_name")[["reading_score","math_score"]].mean().reset_index()
avg_test_scores_df


```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_maths=school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name').size().reset_index(name='above_70_math')
school_passing_maths

```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name').size().reset_index(name = 'above_70_reading')
school_passing_reading

```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)].groupby("school_name").size().reset_index(name ='passing_both')

passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates

passing_math=school_data_complete.groupby(['school_name']).aggregate({'school_name':'count'})
passing_math.columns=['Total_student']


per_school_passing_math=pd.merge(school_passing_maths,passing_math,how="left",on=["school_name"])
per_school_passing_math["%Passing Math"]=per_school_passing_math['above_70_math']/per_school_passing_math['Total_student']*100
per_school_passing_math_df=per_school_passing_math.drop(columns=['above_70_math','Total_student'])
per_school_passing_math_df
```

```python
passing_reading=school_data_complete.groupby(['school_name']).aggregate({'school_name':'count'})
passing_reading.columns=['Total_student']


per_school_passing_reading=pd.merge(school_passing_reading,passing_reading,how="left",on=["school_name"])
per_school_passing_reading["%Passing Reading"]=per_school_passing_reading['above_70_reading']/per_school_passing_reading['Total_student']*100
per_school_passing_reading_df=per_school_passing_reading.drop(columns=['above_70_reading','Total_student'])

per_school_passing_reading_df
```

```python
# Use the provided code to calculate the passing rates

passing_df=school_data_complete.groupby(['school_name']).aggregate({'school_name':'count'})

passing_df.columns=['Total_student']

passing_rates_df=pd.merge(passing_math_and_reading, passing_df, how="left", on=["school_name"])
passing_rates_df["percentage-passed"]=passing_rates_df['passing_both']/passing_rates_df['Total_student']*100
passing_rates_df

```

```python
total_budget = school_data[['school_name','budget']]
budget_df=pd.DataFrame(total_budget)
budget_df
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.




per_school_summary=passing_rates_df
per_school_summary=pd.merge(per_school_summary,school_types, how="left", on=["school_name"])
# per_school_summary
per_school_summary=pd.merge(per_school_summary,avg_test_scores_df, how="left", on=["school_name"])
# per_school_summary
per_school_summary=pd.merge(per_school_summary,per_school_passing_math_df, how="left", on=["school_name"])
# per_school_summary
per_school_summary=pd.merge(per_school_summary,per_school_passing_reading_df, how="left", on=["school_name"])
per_school_summary=pd.merge(per_school_summary,per_school_capita[['school_name','per_capita']],on=["school_name"], how="left")
# per_school_summary=pd.merge(per_school_summary,df, how="left", on=["school_name"])
per_school_summary=pd.merge(per_school_summary,budget_df,how="left",on=["school_name"])
per_school_summary



# # Formatting
per_school_summary["budget"] = per_school_summary["budget"].map("${:,.2f}".format)
per_school_summary["per_capita"] = per_school_summary["per_capita"].map("${:,.2f}".format)


# Display the DataFrame
per_school_summary

```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by "school_name" and take the mean of each.
ninth_graders_scores = ninth_graders.groupby(['school_name'])['math_score'].mean().reset_index(name="9th_mathscore")
tenth_graders_scores = tenth_graders.groupby(['school_name'])['math_score'].mean().reset_index(name="10th_mathscore")
eleventh_graders_scores = eleventh_graders.groupby(['school_name'])['math_score'].mean().reset_index(name="11th_mathscore")
twelfth_graders_scores = twelfth_graders.groupby(['school_name'])['math_score'].mean().reset_index(name="12th_mathscore")



# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade=pd.merge(ninth_graders_scores,tenth_graders_scores, how="left", on=["school_name"])
math_scores_by_grade
math_scores_by_grade=pd.merge(math_scores_by_grade,eleventh_graders_scores, how="left", on=["school_name"])
math_scores_by_grade=pd.merge(math_scores_by_grade,twelfth_graders_scores, how="left", on=["school_name"])




math_scores_by_grade.head()
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by "school_name" and take the mean of each.

ninth_graders.reset_index(drop=True)
ninth_graders.groupby('school_name').mean()['reading_score']
ninth_graders_scores = ninth_graders.groupby(['school_name'])['reading_score'].mean().reset_index(name="9th_readingscore")
ninth_graders_scores
tenth_graders_scores = tenth_graders.groupby(['school_name'])['reading_score'].mean().reset_index(name="10th_readingscore")
tenth_graders_scores
eleventh_graders_scores = eleventh_graders.groupby(['school_name'])['reading_score'].mean().reset_index(name="11th_readingscore")
eleventh_graders_scores
twelfth_graders_scores = twelfth_graders.groupby(['school_name'])['reading_score'].mean().reset_index(name="12th_readingscore")
twelfth_graders_scores


# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade=pd.merge(ninth_graders_scores,tenth_graders_scores, how="left", on=["school_name"])
reading_scores_by_grade=pd.merge(reading_scores_by_grade,eleventh_graders_scores, how="left", on=["school_name"])
reading_scores_by_grade=pd.merge(reading_scores_by_grade,twelfth_graders_scores, how="left", on=["school_name"])


# Display the DataFrame
reading_scores_by_grade.head()
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df

```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['per_capita'].str.replace('$','').astype(float),bins=spending_bins,labels=labels)

school_spending_df.reset_index(drop=True)

```

```python
new_school_spending_df = school_spending_df.rename(columns={'passing_both':'# Students Passing Both','math_score':'Average Math Score','reading_score':'Average Reading Score','above_70_math':'%Passing Math','above_70_reading':'%Passing Reading','percentage-passed':'%Overall Passing','Total_student':'Total Students','school_type':'School Type','per_capita':'Per Student Budget','budget':'Total School Budget'})


new_spending_index = new_school_spending_df.reset_index(drop=True)
new_spending_index
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = new_school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = new_school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = new_school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["%Passing Math"]
spending_passing_reading = new_school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["%Passing Reading"]
overall_passing_spending = new_school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["%Overall Passing"]


```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({'Average Math Scores':spending_math_scores.values,'Average Reading Scores':spending_reading_scores.values,
                                 '%Passing Math':spending_passing_math.values,'%Passing Reading':spending_passing_reading.values,
                                 '%Overall Passing':overall_passing_spending.values,'Spending Ranges (Per Student)':["<$585", "$585-630", "$630-645", "$645-680"]
                                  })


spending_summary.set_index('Spending Ranges (Per Student)')



```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

new_spending_index['School Size']=pd.cut(new_spending_index['Total Students'],bins=size_bins,labels=labels)
new_spending_index
```

```python
# Calculate averages for the desired columns. 
size_math_scores = new_spending_index.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = new_spending_index.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = new_spending_index.groupby(["School Size"]).mean()["%Passing Math"]
size_passing_reading = new_spending_index.groupby(["School Size"]).mean()["%Passing Reading"]
size_overall_passing = new_spending_index.groupby(["School Size"]).mean()["%Overall Passing"]


```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({'Average Math Scores':size_math_scores.values,'Average Reading Scores':size_reading_scores.values,
                             '%Passing Math':size_passing_math.values,'%Passing Reading':size_passing_reading.values,
                             '%Overall Passing':size_overall_passing.values,'School Size':["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]}) 

size_summary.set_index('School Size')

# Display results
size_summary

```
